```python
import pandas as pd
```

```python
df = pd.read_excel("df_MLReady.xlsx")
df.head()
```

```python
#set const
max_change = 0.1
```

```python
#df["Price_Change_Normalized"] = max(min((df["Price_Prior"] - df["Price_After"]) / df["Price_Prior"],-max_change),max_change)
df["Price_Change_Normalized(LABEL)"] = (df["Price_After"] - df["Price_Prior"]) / df["Price_Prior"] / max_change 
df["Price_Change_Normalized(LABEL)"] = df["Price_Change_Normalized(LABEL)"].clip(-1,1)
df["Price_DURING_Normalized(INPUT)"] = (df["Price_During"] - df["Price_Prior"]) / df["Price_Prior"] / max_change 
df["Price_DURING_Normalized(INPUT)"] = df["Price_DURING_Normalized(INPUT)"].clip(-1,1)
df
```

```python
df.to_excel("df_MLReady_Normalized_V2.xlsx")
```
