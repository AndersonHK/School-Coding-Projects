```python
import pandas as pd
```

It's important to set the separator to tabs, that's \t.

```python
df = pd.read_csv("df_MLReady_Normalized_V2_Merged.csv", sep ='\t')
df.head()
```

When you first read the file, Pandas doesn't figure that the text cells are an array.

```python
df["Text"][0]
```

```python
len(df["Text"][0])
```

To fix that, you got to manually replace the commas and quotes with a delimiter, in this case I'm using /t again.

```python
split_ready = df
split_ready["Text"] = split_ready["Text"].apply(lambda x: x[2:-2].replace('", "','/t')) #.replace('"/t"','/t')
split_ready["Text"][0]
```

Now that we have the delimiters, we can split the text into the different earning calls.

```python
split_ready["Text"] = split_ready["Text"].apply(lambda x: x.split('/t'))
split_ready["Text"][0]
```
